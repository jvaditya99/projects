```python
import sympy as sp
from sympy import Symbol, exp, log, E, sin, cos, pi
from IPython.core.display import Markdown
import numpy as np
from numpy import linalg as LA
from sympy.calculus.util import *
import warnings
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
warnings.filterwarnings("ignore", category=np.VisibleDeprecationWarning)


n = int(input('Enter the number of variables(2,3)'))
g=Symbol('g')

if n == 2:
    x1=Symbol('x\u2081')
    x2=Symbol('x\u2082')
    f1=Symbol('f\u2081')
    f2=Symbol('f\u2082')
    qn = input('Enter the question number with 2 variables(1a,1b,1c,1d)')
    if qn == '1a':
        #1a
        f1 = 4*x1**2 -20*x1 + 0.25*x2**2 + 8
        f2 = 0.5* x1*x2**2 + 2*x1 -5*x2 + 8
        x0 = np.array([0,0]).reshape((n, 1))
    elif qn == '1b':
        #1b
        f1 = 3 *(x1**2)-(x2**2)
        f2 = 3*x1*x2 - x1**3 -1
        x0 = np.array([1,1]).reshape((n, 1))
    elif qn == '1c':
        #1c
        f1 = log((x1**2)+(x2**2)) - sin(x1*x2) - log(2) - log(pi)
        f2 = exp(x1-x2) + cos(x1*x2)
        x0 = np.array([1,1]).reshape((n, 1))
    elif qn == '1d':
        #1d
        f1 = sin(4*pi*x1*x2) - 2*x2 -x1
        f2 = ((4*pi - 1)/(4*pi)) * (exp(2*x1)-exp(1)) + 4*exp(1)*x2**2 - 2*exp(1)*x1
        x0 = np.array([0,0]).reshape((n, 1))
    else:
        print('Invalid question')
    
    g = f1**2 + f2**2
    variables = (x1,x2)
    functions = np.array([f1,f2]).reshape((n, 1))
    
elif n == 3:
    x1=Symbol('x\u2081')
    x2=Symbol('x\u2082')
    x3=Symbol('x\u2083')
    f1=Symbol('f\u2081')
    f2=Symbol('f\u2082')
    f3=Symbol('f\u2083')
    qn = input('Enter the question number with 3 variables(example1,2a,2b,2c,2d)')
    if qn == 'example1':
        #example1
        f1 = 3 * x1 - cos(x2*x3) -0.5
        f2 = x1**2 - 81 *(x2 + 0.1)**2 + sin(x3) + 1.06
        f3 = exp(-x1*x2) + 20*x3 + (10*pi -3)/3
        x0 = np.array([0,0,0]).reshape((n, 1))
    elif qn == '2a':
        #2a
        f1 = 15*x1 + x2**2 - 4*x3 - 13
        f2 = x1**2 + 10*x2 -x3 - 11
        f3 = x2**3 -25*x3 + 22
        x0 = np.array([0,0,0]).reshape((n, 1))
    elif qn == '2b':
        #2b
        f1 = 10 *x1 -2*x2**2 + x2 -2*x3 - 5
        f2 = 8*x2**2 + 4*x3**2 - 9
        f3 = 8*x2*x3 + 4
        x0 = np.array([0,0,0]).reshape((n, 1))
    elif qn == '2c':
        #2c
        f1 = x1**3 + x2*x1**2 - x1*x3 +6
        f2 = exp(x1) + exp(x2) -x3
        f3 = x2**2 - 2*x1*x3 - 4 
        x0 = np.array([0,0,0]).reshape((n, 1))
    elif qn == '2d':
        #2d
        f1 = x1 + cos(x1*x2*x3) - 1
        f2 = (1 - x1)**0.25 + x2 + 0.05*x3**2 - 0.15*x3 - 1
        f3 = -x1**2 - 0.1*x2**2 + 0.01*x2 + x3 -1
        x0 = np.array([0,0,0]).reshape((n, 1))
    else:
        print('Invalid question')
         
    g = f1**2 + f2**2 + f3**2
    variables = (x1,x2,x3)
    functions = np.array([f1,f2,f3]).reshape((n, 1))
else:
    print('Invalid number')


def Jacobian(functions,variables):
    var = variables
    f = list(functions)
    Ja = sp.zeros(len(variables),len(functions))
    for i, fi in enumerate(f):
        for j, s in enumerate(var):
            Ja[i,j] = fi[0].diff(s,1)

    return np.array(Ja).astype(sp.core.add.Add)


def SubstitiueVectorInEquation(g,x0):
    con = g
    for i in range(len(x0)):
        con = con.subs(variables[i],x0[i][0])
    return float(con)  



def SubstitiueVectorInMatrix(mat,x0):
    lst = np.zeros(mat.shape)
    for i,fi in enumerate(mat):
        for j,ki in enumerate(fi):
            lst[i][j] = SubstitiueVectorInEquation(ki,x0)
        
    return lst


xnew = x0
gold = 1
gnew = 0
X = 0
while (gold - gnew) > 0.05:
    gradient_vector = np.matmul(2*np.transpose(Jacobian(functions,variables)),(functions)).astype(sp.core.add.Add)
    z = (SubstitiueVectorInMatrix(gradient_vector,xnew)/LA.norm(SubstitiueVectorInMatrix(gradient_vector,xnew)))
    
    ################### to find alpha #############################
    al1 = 0
    al3 = 1
    while SubstitiueVectorInEquation(g,xnew-al3*z) >= SubstitiueVectorInEquation(g,xnew-al1*z):
        al3 = al3/2
    al2 = al3/2
    g1 =  SubstitiueVectorInEquation(g,xnew)
    g2 = SubstitiueVectorInEquation(g,xnew-al2*z)
    g3 = SubstitiueVectorInEquation(g,xnew-al3*z)
    h1 = (g2-g1)/al2
    h2 = (g3 - g2)/(al3 - al2)
    h3 = (h2 -h1)/al3
    al0 = 0.5 * (al2 - (h1/h3))
    g0 = SubstitiueVectorInEquation(g,x0-al0*z)   
    dic  = {al0:g0,al1:g1,al2:g2,al3:g3,}
    al = min(dic, key = lambda a : dic[a])
    ###############################################################
    
    xold = xnew
    gold = SubstitiueVectorInEquation(g,xold)
    print(f'X{X} = ',np.transpose(xold),f'g(X{X}) = ',gold)
    xnew = xnew-al*z
    gnew = SubstitiueVectorInEquation(g,xnew)
    X+=1
    

```
